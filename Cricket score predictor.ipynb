```python
df["current_score"] = df.groupby("match_id")[["Unnamed: 0","ball","runs"]].cumsum()["runs"]
```

```python
df.head()
```

```python
df["over"] = df["ball"].apply(lambda x: str(x).split(".")[0])
df
```

```python
df["ball_no."] = df["ball"].apply(lambda x: str(x).split(".")[1])
df
```

```python
df["ball_bowled"] = (df["over"].astype(int)*6+df["ball_no."].astype(int))
df.tail()
```

```python
# the ball should be 120 but in the 63887 it is 121 so we have to handle this 
```

```python
df["ball_left"] = 120-df["ball_bowled"]
```

```python
df
```

```python
df["ball_left"] = df["ball_left"].apply(lambda x:0 if x<0 else x)
df
```

```python
df.tail()
```

```python
df.info()
```

```python
df["player_dismissed"] = df["player_dismissed"].apply(lambda x:1 if x!="0" else 0)
```

```python
df
```

```python
df["player_dismissed"] = df["player_dismissed"].astype(int)
```

```python
df["player_dismissed"] = df.groupby("match_id")[["player_dismissed"]].cumsum()["player_dismissed"]
```

```python
df
```

```python
df["wicket_left"] = 10-df["player_dismissed"]
df
```

```python
df["current_run_rate"] = df["current_score"]*6/df["ball_bowled"]
df
```

```python
groups = df.groupby("match_id")[["runs"]]
#5 over = 30ball
match_id = df["match_id"].unique()
last_five = []
for id in match_id:
    last_five.extend(groups.get_group(id).rolling(window = 30).sum()["runs"].values.tolist())
    
```

```python
last_five
```

```python
df["last_five"] = last_five
df
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error

# Define which columns are numeric and which are categorical
numeric_features = ['current_score', 'ball_left', 'wicket_left', 'current_run_rate', 'last_five']  # Replace with your actual numeric columns
categorical_features = ['batting_team', 'bowling_team', 'city']  # Replace with your actual categorical columns

# Define the preprocessing steps for numeric and categorical data
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Combine the preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

# Create the full pipeline with the preprocessor and the regressor
pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1))])

# Fit the pipeline on the training data
pipe.fit(X_train, y_train)

# Predict on the test data
y_pred = pipe.predict(X_test)

# Evaluate the model
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f'R^2 Score: {r2}')
print(f'Mean Absolute Error: {mae}')
```
